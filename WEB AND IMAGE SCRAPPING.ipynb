```python
import urllib.request
```

```python
import random  #to ensure all the file names are random and not same.
```

```python
for link in n:
    href=link.get('href')
    print(href)
    img_name=random.randrange(1,500)
    full_name=str(img_name)+".jpg"
    urllib.request.urlretrieve(href,full_name)
    print("scraping")
    print(full_name)
```

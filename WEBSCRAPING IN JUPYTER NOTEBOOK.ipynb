```python
from autoscraper import AutoScraper
```

```python
amazon_url="https://www.amazon.in/s?k=iphone"##gave url input
wanted_list=["₹69,900","Apple iPhone 13 (128GB) - Midnight"]
```

```python
scraper=AutoScraper()
result=scraper.build(amazon_url,wanted_list)##for scraping wanted_list from given url.
print(result)
```

```python
scraper.get_result_similar(amazon_url,grouped=True)##grouped as we take 3 important info as present in wanted list.
```

```python
scraper.set_rule_aliases({'rule_0xzz':'Title','rule_s0bn':'Price'})##rule name may change after every execution.# rule name contains info of product
scraper.keep_rules(['rule_0xzz','rule_s0bn'])##to be applied on other pages as well
scraper.save('amazon-search')
```

```python
results=scraper.get_result_similar('https://www.amazon.in/s?k=mi+phones+under+15000',group_by_alias=True)##group_by_alias=True so that
##whatever grouping technique we have used same will be used also.
```

```python
results['Title']##as named earlier Title
```
